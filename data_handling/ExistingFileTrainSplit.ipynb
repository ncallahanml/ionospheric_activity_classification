# Quick and Dirty File Location Parsing

```python
import os
import pandas as pd
from shutil import copy
from random import shuffle

lucas_filepath = 'augmented_B/'
nick_filepath = 'augmented_A/'

nick_files = os.listdir(nick_filepath)
nick_xml_paths = [file for file in nick_files if file.endswith('.xml')]
nick_jpg_paths = [file for file in nick_files if file.endswith('.jpg')]
nick_csv_paths = [file for file in nick_files if file.endswith('.csv')]

lucas_files = os.listdir(lucas_filepath)
lucas_xml_paths = [file for file in lucas_files if file.endswith('.xml')]
lucas_jpg_paths = [file for file in lucas_files if file.endswith('.jpg')]
lucas_csv_paths = [file for file in lucas_files if file.endswith('.csv')]

nick_paired_list = list(zip(nick_xml_paths, nick_jpg_paths, nick_csv_paths))
lucas_paired_list = list(zip(lucas_xml_paths, lucas_jpg_paths, lucas_csv_paths))
```

```python
shuffle(nick_paired_list)
shuffle(lucas_paired_list)

print(len(nick_paired_list))
print(len(lucas_paired_list))
```

```python
VALID_SPLIT = .1
TRAIN_SPLIT = .7
TEST_SPLIT = .2

valid_len = int(len(nick_paired_list) * VALID_SPLIT)
train_len = int(len(nick_paired_list) * TRAIN_SPLIT)
# test_len = int(len(nick_paired_list) * TEST_SPLIT)

nick_xmlpath_list = list()
nick_type_list = list()
nick_jpgpath_list = list()
nick_csvpath_list = list()

lucas_xmlpath_list = list()
lucas_type_list = list()
lucas_jpgpath_list = list()
lucas_csvpath_list = list()

directories = [
'nick_train/',
'nick_test/',
'nick_valid/',

'lucas_train/',
'lucas_test/',
'lucas_valid/',
]
for directory in directories:
    if not os.path.exists(directory):
        os.mkdir(directory)
```

```python
for i, (xml_path, jpg_path, csv_path) in enumerate(nick_paired_list):
    print(xml_path)
    print(jpg_path)
    if i < valid_len:
        new_directory = directories[5]
        nick_type_list.append('VALID')

#         new_filename = new_directory + xml_path.split('/')[-1]
#         copy(nick_filepath + xml_path, new_filename)
#         nick_xmlpath_list.append(new_filename)
        
#         new_filename = new_directory + jpg_path.split('/')[-1]
#         copy(nick_filepath + jpg_path, new_filename)
#         nick_jpgpath_list.append(new_filename)
        
#         new_filename = new_directory + csv_path.split('/')[-1]
#         copy(nick_filepath + csv_path, new_filename)
#         nick_jpgpath_list.append(new_filename)
        
    elif i < valid_len + train_len:
        new_directory = directories[3]
        nick_type_list.append('TRAIN')

#         new_filename = new_directory + xml_path.split('/')[-1]
#         copy(nick_filepath + xml_path, new_filename)
#         nick_xmlpath_list.append(new_filename)
        
#         new_filename = new_directory + jpg_path.split('/')[-1]
#         copy(nick_filepath + jpg_path, new_filename)
#         nick_jpgpath_list.append(new_filename)
        
#         new_filename = new_directory + csv_path.split('/')[-1]
#         copy(nick_filepath + csv_path, new_filename)
#         nick_csvpath_list.append(new_filename)
        
    else:
        new_directory = directories[4]
        nick_type_list.append('TEST')
        
    new_filename = new_directory + xml_path.split('/')[-1]
    copy(nick_filepath + xml_path, new_filename)
    nick_xmlpath_list.append(new_filename)

    new_filename = new_directory + jpg_path.split('/')[-1]
    copy(nick_filepath + jpg_path, new_filename)
    nick_jpgpath_list.append(new_filename)

    new_filename = new_directory + csv_path.split('/')[-1]
    copy(nick_filepath + csv_path, new_filename)
    nick_csvpath_list.append(new_filename)
        
        
for i, (xml_path, jpg_path, csv_path) in enumerate(lucas_paired_list):
    if i < valid_len:
        new_directory = directories[2]
        lucas_type_list.append('VALID')

#         new_filename = new_directory + xml_path.split('/')[-1]
#         copy(lucas_filepath + xml_path, new_filename)
#         lucas_xmlpath_list.append(new_filename)
#         new_filename = new_directory + jpg_path.split('/')[-1]
#         lucas_jpgpath_list.append(new_filename)
#         copy(lucas_filepath + jpg_path, new_filename)
    elif i < valid_len + train_len:
        new_directory = directories[0]
        lucas_type_list.append('TRAIN')

#         new_filename = new_directory + xml_path.split('/')[-1]
#         copy(lucas_filepath + xml_path, new_filename)
#         lucas_xmlpath_list.append(new_filename)
#         new_filename = new_directory + jpg_path.split('/')[-1]
#         lucas_jpgpath_list.append(new_filename)
#         copy(lucas_filepath + jpg_path, new_filename)
    else:
        new_directory = directories[1]
        lucas_type_list.append('TEST')

#         new_filename = new_directory + xml_path.split('/')[-1]
#         copy(lucas_filepath + xml_path, new_filename)
#         lucas_xmlpath_list.append(new_filename)
#         new_filename = new_directory + jpg_path.split('/')[-1]
#         lucas_jpgpath_list.append(new_filename)
#         copy(lucas_filepath + jpg_path, new_filename)
    new_filename = new_directory + xml_path.split('/')[-1]
    copy(lucas_filepath + xml_path, new_filename)
    lucas_xmlpath_list.append(new_filename)

    new_filename = new_directory + jpg_path.split('/')[-1]
    copy(lucas_filepath + jpg_path, new_filename)
    lucas_jpgpath_list.append(new_filename)

    new_filename = new_directory + csv_path.split('/')[-1]
    copy(lucas_filepath + csv_path, new_filename)
    lucas_csvpath_list.append(new_filename)
print('complete')
```

```python
df = pd.DataFrame()
df['Nick XML'] = nick_xmlpath_list
df['Nick JPG'] = nick_jpgpath_list
df['Nick Split'] = nick_type_list

df['Lucas XML'] = lucas_xmlpath_list
df['Lucas JPG'] = lucas_jpgpath_list
df['Lucas Split'] = lucas_type_list

display(df)
df.to_csv('attribution_df.csv')
```
