```python
def csv_adjust(directory, method='test'):
    files = [directory + '/' + file for file in os.listdir(directory) if file.endswith('.csv')]
    df_list = list()
    for file in files:
        temp_df = pd.read_csv(file, index_col=0)
        df_list.append(temp_df)
    
    
#     f = lambda file : file.split('/')[-1].split('_')[0] + 
    
    
    df = pd.concat(df_list, axis=0)
    df['filename'] = df['filename'].apply(f)
    display(df)
    df = df.drop_duplicates()
    df.to_csv(f'{method}.csv', index=None)
    return df

def rename_images(directory):
    files = [directory + '/' + file for file in os.listdir(directory) if file.endswith('.jpg')]
    for file in files:
        new_file = directory + adj_file_name(file)
        move(file, new_file)
    return
```
